```python
#JAMIL MAHMOOD
#ROLL : 2017331084

import numpy as np
import matplotlib.pyplot as plt
import math

def plot_signal(t, x):

  plt.figure(figsize=(15, 4))
  plt.plot(t, x)
  plt.show()

  #Amplitude Modulation(AM)

#Step 1: Creating Carrier Signal

A=2; 
fc=0.5; 
t=np.arange(0, 50, 0.1)
ct=A*np.cos(2*np.pi*fc*t); 
plot_signal(t, ct)

```

```python
#Step 2: Creating message Signal

m = 1.0
M = A * m
fm=.05
phase=0
mt=M*np.sin((2*np.pi*fm*t)+phase)
plot_signal(t, mt)
```

```python
#Step 3: Computing Amplitude modulation

yt=(1+m*np.sin((2*np.pi*fm*t) + phase))*A*np.cos(2*np.pi*fc*t);
plot_signal(t, yt)
```

```python
#Frequency Modulation(FM)

#Step 1: Creating Carrier Signal

Ac=2; 
fc=25; 
t=np.arange(0, 1.0, 0.0001)
xc=Ac*np.sin(2*np.pi*fc*t); 
plot_signal(t, xc)
```

```python
#Step 2: Creating modulating signal

m = 1.
fm=4
xm=np.cos((2*np.pi*fm*t))
plot_signal(t, xm)
```

```python
#Step 3: Computing Frequency modulation

yt = Ac*np.sin(2. * np.pi * (fc * t + m * np.cos(2*np.pi*fm*t)))
plot_signal(t, yt)
```

```python
#Phase Modulation(PM)

#Step 1: Creating Carrier Signal

Ac=2; 
fc=2; 
t=np.arange(0, 3.0, 0.0001)
phase = 0
ct=Ac*np.cos(2*np.pi*fc*t + phase); 
plot_signal(t, ct)
```

```python
#Step 2: Creating message Signal

k = 1.0
fm=3
Am = 2
mt=Am*np.cos(2 * np.pi*fm*t)
plot_signal(t, mt)
```

```python
#Step 3: Computing Phase modulation

yt=Ac*np.cos((2*np.pi*fc*t) + (k * mt + phase))
plot_signal(t, yt)
```
