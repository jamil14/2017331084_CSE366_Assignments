```python
#JAMIL MAHMOOD
#ROLL : 2017331084

#Different subplots of a sinusoidal signal

import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 2.0, 80)
def create_sinusoidal(A, f, phase = 0):
  x = A * np.cos(2.0 * np.pi * f * t + phase)
  return x
x = create_sinusoidal(1,1)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(13, 11))
fig.suptitle('Different subplots of a sinusoidal signal', fontsize=20)

ax[0, 0].plot(t,x)
ax[0, 0].set_title('continuous plot', fontsize = 15)

ax[0, 1].stem(t,x)
ax[0, 1].set_title('stem plot', fontsize = 15)

ax[1,0].scatter(t,x)
ax[1,0].set_title('scatter plot', fontsize = 15)
ax[-1,-1].axis('off')

# for a in ax.flat:
#     a.set(xlabel='t', ylabel='x(t)')

plt.show()
```

```python
#3 Sinusoidal signal and a combined signal
x1 = create_sinusoidal(1,1)
x2 = create_sinusoidal(4,4)
x3 = create_sinusoidal(5,5)
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(13, 11))
fig.suptitle('3 Sinusoidal signal and a combined signal', fontsize=20)

ax[0, 0].plot(t,x1, 'r')
ax[0, 0].set_title('x1(t)', fontsize = 15)

ax[0, 1].plot(t,x2, 'g')
ax[0, 1].set_title('x2(t)', fontsize = 15)

ax[1,0].plot(t,x3, 'b')
ax[1,0].set_title('x3(t)', fontsize = 15)

ax[1,1].plot(t, x1 + x2 + x3)
ax[1,1].set_title('y(t) = x1(t) + x2(t) + x3(t)', fontsize = 15)
plt.show()
```
