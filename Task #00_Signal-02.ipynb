```python
#JAMIL MAHMOOD
#ROLL : 2017331084

#3 Sinusoidal Signals and 1 Composite Signal

import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.0,2.0,0.00001)
def create_sinusoidal(t, A, f, phase = 0):
  x = A * np.cos(2.0 * np.pi* f * t + phase)
  return x
x1 = create_sinusoidal(t, 5, 5)
x2 = create_sinusoidal(t, 2, 2)
x3 = create_sinusoidal(t, 1, 10)
y = x1 + x2 + x3
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(13, 11))
fig.suptitle('3 Sinusoidal Signals and 1 Composite Signal', fontsize=20)
ax[0,0].plot(t, x1)
ax[0,0].set_title('Signal-1', fontsize = 15)

ax[0,1].plot(t, x2)
ax[0,1].set_title('Signal-2', fontsize = 15)

ax[1,0].plot(t, x3)
ax[1,0].set_title('Signal-3', fontsize = 15)

ax[1,1].plot(t, y)
ax[1,1].set_title('Signal-1 + Signal-2 + Signal-3', fontsize = 15)

plt.show()
```

```python
fs = 10
step = 1.0/fs
n = np.arange(0, 2.0, step)
y = create_sinusoidal(n, 5, 5) + create_sinusoidal(n, 2, 2) + create_sinusoidal(n, 1, 10)
plt.title('fs = 10', fontsize = 15)
plt.stem(n,y)
plt.show()

fs = 25
step = 1.0/fs
n = np.arange(0, 2.0, step)
y = create_sinusoidal(n, 5, 5) + create_sinusoidal(n, 2, 2) + create_sinusoidal(n, 1, 10)
plt.title('fs = 25', fontsize = 15)
plt.stem(n,y)
plt.show()

fs = 50
step = 1.0/fs
n = np.arange(0, 2.0, step)
y = create_sinusoidal(n, 5, 5) + create_sinusoidal(n, 2, 2) + create_sinusoidal(n, 1, 10)
plt.title('fs = 50', fontsize = 15)
plt.stem(n,y)

plt.show()
```
